```python
import seaborn as sns
import matplotlib.pyplot as plt

for feature in df_train.columns:
    # only show columns == WetlandLoss
    if 'WetlandLoss' in feature:
        plt.figure(figsize=(10,6))
        sns.boxplot(x=df_train[feature])
        plt.title(f'Box Plot of {feature}')
        plt.show()
```

```python
# identify the outliers using the IQR method
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3 - Q1

# filter dataframe for save line which are not outliers in numeric columns
condition = ~((df_train < (Q1 - 1.5 * IQR)) | (df_train > (Q3 + 1.5 * IQR))).any(axis=1)
df = df_train.loc[condition, df_train.columns]
```

```python
for feature in df.columns:
    # only show columns == WetlandLoss
    if 'WetlandLoss' in feature:
        plt.figure(figsize=(10,6))
        sns.boxplot(x=df[feature])
        plt.title(f'Box Plot of {feature}')
        plt.show()
```

```python
for feature in df_train.columns:
    if 'Deforestation' in feature:
        plt.figure(figsize=(10,6))
        sns.histplot(x=df_train[feature], kde=True)
        plt.title(f'Box Plot of {feature}')
        plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# make sure only numeric columns are processed
numeric_features = df.select_dtypes(include=['number']).columns
numeric_features

# standardize numeric feature
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
for feature in df.columns:
    if 'Deforestation' in feature:
        plt.figure(figsize=(10,6))
        sns.histplot(x=df[feature], kde=True)
        plt.title(f'Box Plot of {feature}')
        plt.show()
```

```python
# count total of variables
num_vars = df.shape[1]

# determine total of rows and columns for grid subplot
n_cols = 4 # desired number of columns
n_rows = -(-num_vars // n_cols) # ceiling division for determine number of rows

# create subplot
fig, ax = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))

# flatten ax array for easy to iterate if needed
ax = ax.flatten()

# plot every variable
for i, column in enumerate(df.columns):
  df[column].hist(ax=ax[i], bins=20, edgecolor='black')
  ax[i].set_title(column)
  ax[i].set_xlabel('Value')
  ax[i].set_ylabel('Frequency')
  
# remove subplot that not used (if any)
for j in range(i+1, len(ax)):
  fig.delaxes(ax[j])
  
# adjust layout to make it neater
plt.tight_layout()
plt.show()
```

```python
# count corelation between traget variable and all other variables
target_corr = df.corr()['FloodProbability']

# (optional) sort the correlation value by power of the correlation
target_corr_sorted = target_corr.abs().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
target_corr_sorted.plot(kind='bar')
plt.title('Correlation with SalePrice')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
import sklearn
from sklearn import datasets

# seperate x feature and y target
X = df.drop(columns=['FloodProbability'])
y = df['FloodProbability']

from sklearn.model_selection import train_test_split

# split data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# count long/total data
print(f'Total data: {len(X)}')

# count long/total data for train and test set
print(f'Total data for train set: {len(X_train)}')
print(f'Total data for test set: {len(X_test)}')
```

```python
# LARS
from sklearn import linear_model
lars = linear_model.Lars(n_nonzero_coefs=1).fit(X_train, y_train)

pred_lars =lars.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)

print(f'MAE: {mae_lars}')
print(f'MSE: {mse_lars}')
print(f'R2: {r2_lars}')
```

```python
# create dictionary for save the result
data = {
    'Model': ['LARS'],
    'MAE': [mae_lars],
    'MSE': [mse_lars],
    'R2': [r2_lars]
}

# convert dictionary to dataframe
df_results = pd.DataFrame(data, index=['Lars'])
df_results
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
LR = LinearRegression().fit(X_train, y_train)

pred_LR = LR.predict(X_test)

mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)

print(f'MAE: {mae_LR}')
print(f'MSE: {mse_LR}')
print(f'R2: {r2_LR}')
```

```python
lr = 'Linear Regression'

df_results.loc[lr] = [lr, mae_LR, mse_LR, r2_LR]
df_results
```

```python
# Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor

GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(X_train, y_train)

pred_GBR = GBR.predict(X_test)

mae_GBR = mean_absolute_error(y_test, pred_GBR)
mse_GBR = mean_squared_error(y_test, pred_GBR)
r2_GBR = r2_score(y_test, pred_GBR)

print(f'MAE: {mae_GBR}')
print(f'MSE: {mse_GBR}')
print(f'R2: {r2_GBR}')
```
